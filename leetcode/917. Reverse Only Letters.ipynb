```c
#include <stdio.h>
#include <string.h>
#include <ctype.h>

char r[101];
char e1;
char e2;

char *reverseOnlyLetters(char *S)
{
    strcpy(r, "");
    int length = (int)strlen(S);
    int i1 = 0;
    int i2 = 0;
    int i2Temp = 0;
    //printf("%d", length);
    for (int i = 0; i < length; ++i)
    {
        e1 = S[i];
        if (isalpha(e1) != 0)
        {
            i2Temp = i2;
            while (i2Temp < length)
            {
                e2 = S[length - i2Temp - 1];
                if (isalpha(e2) != 0)
                {
                    strcat(r, &e2);
                    i2 = i2Temp + 1;
                    break;
                }
                i2Temp += 1;
            }
        }
        else
        {
            strcat(r, &e1);
        }
        i1 += 1;
    }
    return r;
}

int main(int argc, char *argv[])
{
    printf("%s", reverseOnlyLetters("hello"));
    printf("%s", "\n\n");
    //printf("%s", "Hello World.\nMy name is yingshaoxo.\n");
}
```
